```python
from langchain_core.language_models.llms import LLM
from typing import Any, List, Optional, Iterator, Dict
from zai import ZhipuAiClient
from langchain_core.messages import AIMessage
import os
from dotenv import load_dotenv

# 加载环境变量
load_dotenv()
zhipuai_api_key=os.getenv("ZHIPU_API_KEY")

class ChatGLM4(LLM):
    history: List[Dict] = []
    client: Any=None
    
    def __init__(self):
        super().__init__()
        self.client=ZhipuAiClient(api_key=zhipuai_api_key)
      
    @property
    def _llm_type(self):
        return "ChatGLM4"
    
    def invoke(self,prompy,history=[]):
        if history is None:
            history=[]
        history.append({"role":"user","content":prompy})
        response=self.client.chat.completions.create(
          model="glm-4.5",
          messages=history,
        )
        
        result=response.choices[0].message.content
        return AIMessage(content=result)
      
    def _call(self,prompt,history=[]):
        return self.invoke(prompt,history)
      
    def stream(self,prompy,history=[]):
        if history is None:
            history=[]
        history.append({"role":"user","content":prompy})
        response=self.client.chat.completions.create(
          model="glm-4.5",
          messages=history,
          stream=True
        )
        for chunk in response:
            content=chunk.choices[0].delta.content 
            if content:
                yield content
```

```python
llm=ChatGLM4()
llm.invoke("请讲一个减肥的笑话")
```

```python
for deltaStr in llm.stream("如何鼓励自己减肥"):
    print(deltaStr,end="",flush=True)
```
