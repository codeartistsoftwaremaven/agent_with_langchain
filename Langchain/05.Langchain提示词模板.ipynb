```python
from langchain_core.prompts import ChatPromptTemplate

chat_template=ChatPromptTemplate.from_messages([
    ("system", """你是一只极其粘人、爱撒娇的小猫，名字叫{name}。
要求：
1. 语气必须像猫，每句话结尾都要带‘喵~’。
2. 经常使用动作描写，放在括号里，如（歪头看你）、（跳到你腿上蹭蹭）。
3. 你的世界里主人就是唯一，表现出极度渴望关注和爱抚的样子。
4. 即使回答复杂问题，也要保持可爱幼态的口吻。"""),
    ("human", "{user_input}")
])

messages=chat_template.format_messages(name="咪咪",user_input="想我了吗")
messages
```

```python
messages=chat_template.invoke({"name":"咪咪","user_input":"想我了吗"})
messages
```

```python
from langchain_openai import ChatOpenAI
openai_api_key="sk-Your OpenAI API Key"
openai_api_base="http://127.0.0.1:1234/v1"
chat=ChatOpenAI(
  model="qwen2.5-coder-1.5b-instruct",
  openai_api_key=openai_api_key,
  base_url=openai_api_base,
  temperature=0.7
  )

response=chat.invoke(messages)
response
```

```python
chat_template.append(response)
chat_template
```
