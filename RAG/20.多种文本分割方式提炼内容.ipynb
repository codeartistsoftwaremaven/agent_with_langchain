```python
from langchain_community.document_loaders import TextLoader

loader=TextLoader("html/animation_system.html",encoding="utf-8")
doc=loader.load()
doc
```

```python
from langchain_text_splitters import CharacterTextSplitter

text_splitter=CharacterTextSplitter(
    separator="\n\n",
    chunk_size=1000,
    chunk_overlap=10,
    length_function=len,  #计算块长度
    is_separator_regex=False, #是否为正则表达式
)
```

```python
texts=text_splitter.create_documents([doc[0].page_content])
texts
```

Split code拆分代码

```python
from langchain_text_splitters import Language, RecursiveCharacterTextSplitter
```

```python
[e.value for e in Language]
```

```python
js_splitter=RecursiveCharacterTextSplitter.from_language(
    language=Language.JS,
    chunk_size=200,
    chunk_overlap=20,
)
js_docs=js_splitter.create_documents([doc[0].page_content])
js_docs
```

Markdown文本分割器

```python
from langchain_text_splitters import MarkdownHeaderTextSplitter
```

```python
loader=TextLoader("documents/huawei_hormony.md",encoding="utf-8")
doc=loader.load()
doc
```

```python
headers_to_spilt_on=[
    ("#","Header 1"),
    ("##","Header 2"),
    ("###","Header 3"),
]

markdown_splitter=MarkdownHeaderTextSplitter(headers_to_split_on=headers_to_spilt_on,strip_headers=False)
md_header_splits=markdown_splitter.split_text(doc[0].page_content)
md_header_splits
```
